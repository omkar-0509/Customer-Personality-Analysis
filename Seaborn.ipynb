```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Customer_Analysis.csv")
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x="Education", y="Income", data=df)
plt.title("Income by Education")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x="Education", data=df)
plt.title("Education Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(y="Marital_Status", data=df)
plt.title("Marital Status Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Income"], bins=30)
plt.title("Income Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.kdeplot(df["Income"], fill=True)
plt.title("Income Density")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x="Income", y="MntWines", data=df)
plt.title("Income vs Wine Spending")
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df[["MntWines","MntFruits","MntMeatProducts"]].head(50))
plt.title("Customer Spending Multiline")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x="Education", y="Income", data=df)
plt.title("Income by Education")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.violinplot(x="Education", y="Income", data=df)
plt.title("Income Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.stripplot(x="Education", y="Income", data=df)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.swarmplot(x="Education", y="Income", data=df)
plt.show()
```

```python
sns.pairplot(df[["Income","MntWines","MntFruits","MntMeatProducts"]])
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.show()
```

```python
sns.jointplot(x="Income", y="MntWines", data=df)
plt.show()
```

```python
sns.regplot(x="Income", y="MntWines", data=df)
plt.show()
```

```python
sns.FacetGrid(df, col="Education").map(sns.histplot, "Income")
plt.show()
```

```python
sns.catplot(x="Education", y="Income", kind="bar", data=df)
plt.show()
```

```python
sns.boxenplot(x="Education", y="Income", data=df)
plt.show()
```

```python
sns.rugplot(df["Income"])
plt.show()
```
