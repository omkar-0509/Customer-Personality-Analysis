```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv("Customer_Analysis.csv")
```

```python
fig = px.scatter_3d(
    df,
    x="Income",
    y="MntWines",
    z="MntMeatProducts",
    color="Education",
    title="3D Scatter Plot - Income vs Wines vs Meat"
)

fig.show()
```

```python
fig = px.scatter_3d(
    df,
    x="MntWines",
    y="MntFruits",
    z="MntMeatProducts",
    color="Marital_Status",
    title="3D Customer Spending"
)

fig.show()
```

```python
fig = px.line_3d(
    df.head(100),
    x="Income",
    y="MntWines",
    z="MntMeatProducts",
    title="3D Line Plot"
)

fig.show()
```

```python
fig = px.scatter_3d(
    df,
    x="Income",
    y="MntWines",
    z="MntMeatProducts",
    size="MntGoldProds",
    color="Education",
    title="3D Bubble Plot"
)

fig.show()
```

```python
pivot = df.pivot_table(
    values="MntWines",
    index="Education",
    columns="Marital_Status",
    aggfunc="mean"
)

fig = go.Figure(
    data=[go.Surface(z=pivot.values)]
)

fig.update_layout(title="3D Surface Plot")

fig.show()
```

```python
fig = go.Figure(data=[
    go.Mesh3d(
        x=df["Income"],
        y=df["MntWines"],
        z=df["MntMeatProducts"],
        opacity=0.5
    )
])

fig.update_layout(title="3D Mesh Plot")

fig.show()
```

```python
fig = px.scatter_3d(
    df,
    x="NumWebPurchases",
    y="NumStorePurchases",
    z="NumCatalogPurchases",
    color="Education",
    title="3D Purchases Plot"
)

fig.show()
```

```python
fig = px.scatter_3d(
    df,
    x="Income",
    y="MntSweetProducts",
    z="MntFishProducts",
    color="Education",
    title="3D Distribution Plot"
)

fig.show()
```

```python
fig = px.scatter_3d(
    df,
    x="Income",
    y="MntSweetProducts",
    z="MntFishProducts",
    color="Education",
    title="3D Distribution Plot"
)

fig.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter3d(
    x=df["Income"],
    y=df["MntWines"],
    z=df["MntMeatProducts"],
    mode="markers",
    name="Spending"
))

fig.show()
```

```python
fig = go.Figure()

fig.add_trace(go.Scatter3d(
    x=df["Income"].head(100),
    y=df["MntWines"].head(100),
    z=df["MntFruits"].head(100),
    mode='lines',
    name='Wines vs Fruits'
))

fig.add_trace(go.Scatter3d(
    x=df["Income"].head(100),
    y=df["MntMeatProducts"].head(100),
    z=df["MntFishProducts"].head(100),
    mode='lines',
    name='Meat vs Fish'
))

fig.update_layout(
    title="3D Multiline Chart - Customer Spending",
    scene=dict(
        xaxis_title='Income',
        yaxis_title='Spending',
        zaxis_title='Products'
    )
)

fig.show()
```

```python
group = df.groupby("Marital_Status")["Income"].mean().reset_index()

fig = go.Figure(data=[go.Scatter3d(
    x=group["Marital_Status"],
    y=[0]*len(group),
    z=group["Income"],
    mode='markers',
    marker=dict(size=12)
)])

fig.update_layout(title="3D Bar Style - Marital Status vs Income")

fig.show()
```

```python
cols = ["MntWines","MntFruits","MntMeatProducts","MntFishProducts"]

values = df[cols].mean()

fig = go.Figure(data=[go.Scatter3d(
    x=cols,
    y=[0]*len(cols),
    z=values,
    mode='markers',
    marker=dict(size=10)
)])

fig.show()
```
