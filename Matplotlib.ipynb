```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Customer_Analysis.csv")
```

```python
plt.figure(figsize=(8,5))
df["Education"].value_counts().plot(kind="bar", color="skyblue")
plt.title("Education Distribution")
plt.xlabel("Education")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["Marital_Status"].value_counts().plot(kind="barh", color="orange")
plt.title("Marital Status Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Income"], bins=30, color="purple", edgecolor="black")
plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(7,7))
df["Education"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%",
    colors=["#ff9999","#66b3ff","#99ff99","#ffcc99"]
)
plt.title("Education Distribution")
plt.ylabel("")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["Year_Birth"].value_counts().sort_index().plot(kind="line", marker="o")
plt.title("Customers by Year of Birth")
plt.xlabel("Year")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Income"], df["MntWines"], alpha=0.5)
plt.title("Income vs Wine Spending")
plt.xlabel("Income")
plt.ylabel("Wine Spending")
plt.show()
```

```python
plt.figure(figsize=(10,5))
df[["MntWines","MntFruits","MntMeatProducts"]].boxplot()
plt.title("Spending Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df[["MntWines","MntFruits","MntMeatProducts"]].head(50).plot(kind="area")
plt.title("Spending Area Plot")
plt.show()
```

```python
df.groupby("Education")[["MntWines","MntFruits"]].mean().plot(
    kind="bar",
    stacked=True,
    figsize=(8,5)
)
plt.title("Stacked Spending by Education")
plt.show()
```

```python
fig, ax = plt.subplots(2,2, figsize=(12,8))

df["Education"].value_counts().plot(kind="bar", ax=ax[0,0])
ax[0,0].set_title("Education")

df["Marital_Status"].value_counts().plot(kind="bar", ax=ax[0,1])
ax[0,1].set_title("Marital Status")

ax[1,0].hist(df["Income"])
ax[1,0].set_title("Income")

ax[1,1].scatter(df["Income"], df["MntWines"])
ax[1,1].set_title("Income vs Wine")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12,8))
corr = df.corr(numeric_only=True)

plt.imshow(corr, cmap="coolwarm")
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation=90)
plt.yticks(range(len(corr)), corr.columns)
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,5))
df["Income"].plot(kind="kde")
plt.title("Income Density Plot")
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.plot(df["MntWines"].head(50), label="Wines")
plt.plot(df["MntFruits"].head(50), label="Fruits")
plt.plot(df["MntMeatProducts"].head(50), label="Meat")
plt.title("Customer Spending Comparison")
plt.xlabel("Customers")
plt.ylabel("Spending")
plt.legend()
plt.show()
```
